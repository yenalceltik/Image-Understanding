```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math

#FOR EDÄ°T OPERATION PRESS RIGHT DOUBLE CLICK

clickPoints=[]#CONTAINS POINTS OF THE RECTANGLE LIKE [X1,Y1,X2,Y2]
editTop=[]#WHEN EDITING I APPEND THE TOP COORDINATES FOR REMINDING
editBottom=[]#WHEN EDITING I APPEND THE BOTTOM COORDINATES FOR REMINDING
edit = False
top = False
controlClick=False
magenta = False

def bound_top(a,b,c,d):#THIS BOUNDS AREAS 2,3,4
    if (a<c and b<d):
        return True
    else:
        return False
        
def bound_bottom(a,b,c,d):#THIS BOUNDS AREAS 1,2,3 
    if a>c and b>d:
        return True
    else:
        return False
        
def control_distance(a,b,c,d):
    distance= math.sqrt((c-a)**2 + (d-b)**2)
    return distance          
        
def draw_rectangle(event,x,y,flags,param):
    global edit, top, img,controlClick, magenta
    img2 =cv2.imread("img2.jpg")
    if event == cv2.EVENT_LBUTTONDBLCLK and edit == False and controlClick==False : 
            magenta = True
            cv2.circle(img,(x,y),3,(0,255,0),-1)
            cv2.circle(img,(x,y),10,(0,255,255),1)
            clickPoints.append(x)
            clickPoints.append(y)

            if len(clickPoints) == 4 and control_distance(clickPoints[0],clickPoints[1],clickPoints[2],clickPoints[3]) > 10:
                
                cv2.rectangle(img,(clickPoints[0],clickPoints[1]),(clickPoints[2],clickPoints[3]),(0,0,255),1)
                controlClick=True  #THIS FLAG FOR,AFTER TWO POINTS SELECTION AND DRAWING RECTANGLE BLOCKING ANY OTHER POINT
                
            elif len(clickPoints) == 4 and control_distance(clickPoints[0],clickPoints[1],clickPoints[2],clickPoints[3]) < 10:
                    img=img2
                    clickPoints.pop(3)
                    clickPoints.pop(2)
                    
                    cv2.circle(img,(clickPoints[0],clickPoints[1]),3,(0,255,0),-1)
                    cv2.circle(img,(clickPoints[0],clickPoints[1]),10,(0,255,255),1)
                                  
    elif event==cv2.EVENT_RBUTTONDBLCLK and controlClick == True:
            
                
            if magenta == True and (abs(clickPoints[0]-x)<10 and abs(clickPoints[1]-y)<10): #DECIDES THE EDITING POINT IS TOP POINT
                img = img2
                
                cv2.circle(img,(clickPoints[0],clickPoints[1]),3,(255,0,255),-1)
                cv2.circle(img,(clickPoints[0],clickPoints[1]),10,(0,255,255),1)
                cv2.circle(img,(clickPoints[2],clickPoints[3]),3,(0,255,0),-1)
                cv2.circle(img,(clickPoints[2],clickPoints[3]),10,(0,255,255),1)
                cv2.rectangle(img,(clickPoints[0],clickPoints[1]),(clickPoints[2],clickPoints[3]),(0,0,255),1)
                editTop.append(clickPoints[0])
                editTop.append(clickPoints[1])
                
                clickPoints.reverse()
                clickPoints.pop(3)
                clickPoints.pop(2)                
                top = True
                edit = True
                magenta = False
            elif magenta == True and (abs(clickPoints[2]-x)<10 and abs(clickPoints[3]-y)<10):#DECIDES THE EDITING POINT IS BOTTOM POINT
                img = img2
                
                cv2.circle(img,(clickPoints[0],clickPoints[1]),3,(0,255,0),-1)
                cv2.circle(img,(clickPoints[0],clickPoints[1]),10,(0,255,255),1)
                cv2.circle(img,(clickPoints[2],clickPoints[3]),3,(255,0,255),-1)
                cv2.circle(img,(clickPoints[2],clickPoints[3]),10,(0,255,255),1)
                cv2.rectangle(img,(clickPoints[0],clickPoints[1]),(clickPoints[2],clickPoints[3]),(0,0,255),1)
                editBottom.append(clickPoints[2])
                editBottom.append(clickPoints[3])
                
                clickPoints.pop(3)
                clickPoints.pop(2)
                top = False
                edit = True
                magenta = False
    elif event==cv2.EVENT_LBUTTONDBLCLK and edit == True and top == True and bound_top(x,y,clickPoints[1],clickPoints[0]) == True:
         
            
            if control_distance(clickPoints[1],clickPoints[0],x,y) > 10:
                img = img2
                clickPoints.append(y)
                clickPoints.append(x)
                clickPoints.reverse()
        
                cv2.circle(img,(x,y),3,(0,255,0),-1)
                cv2.circle(img,(x,y),10,(0,255,255),1)
                cv2.circle(img,(clickPoints[2],clickPoints[3]),3,(0,255,0),-1)
                cv2.circle(img,(clickPoints[2],clickPoints[3]),10,(0,255,255),1)
                cv2.rectangle(img,(clickPoints[0],clickPoints[1]),(clickPoints[2],clickPoints[3]),(0,0,255),1)
                
                edit = False
                magenta = True
            elif control_distance(clickPoints[0],clickPoints[1],x,y) < 10:
                img = img2
               
                
                cv2.circle(img,(editTop[0],editTop[1]),3,(255,0,255),-1)
                cv2.circle(img,(editTop[0],editTop[1]),10,(0,255,255),1)
                cv2.circle(img,(clickPoints[1],clickPoints[0]),3,(0,255,0),-1)
                cv2.circle(img,(clickPoints[1],clickPoints[0]),10,(0,255,255),1)
                cv2.rectangle(img,(editTop[0],editTop[1]),(clickPoints[1],clickPoints[0]),(0,0,255),1)
                
                editTop.pop(1)
                editTop.pop(0)
                      
    elif event==cv2.EVENT_LBUTTONDBLCLK and edit == True and top == False and bound_bottom(x,y,clickPoints[0],clickPoints[1])== True :
           
            
            if  control_distance(clickPoints[0],clickPoints[1],x,y) > 10:
                img = img2
                clickPoints.append(x)
                clickPoints.append(y)
                
                cv2.circle(img,(x,y),3,(0,255,0),-1)
                cv2.circle(img,(x,y),10,(0,255,255),1)
                cv2.circle(img,(clickPoints[0],clickPoints[1]),3,(0,255,0),-1)
                cv2.circle(img,(clickPoints[0],clickPoints[1]),10,(0,255,255),1)  
                cv2.rectangle(img,(clickPoints[0],clickPoints[1]),(clickPoints[2],clickPoints[3]),(0,0,255),1)
                edit = False
                magenta = True
            elif control_distance(clickPoints[1],clickPoints[0],x,y) < 10:
                img = img2
                
                cv2.circle(img,(clickPoints[0],clickPoints[1]),3,(0,255,0),-1)
                cv2.circle(img,(clickPoints[0],clickPoints[1]),10,(0,255,255),1)
                cv2.circle(img,(editBottom[0],editBottom[1]),3,(0,255,0),-1)
                cv2.circle(img,(editBottom[0],editBottom[1]),10,(0,255,255),1)
                cv2.rectangle(img,(clickPoints[0],clickPoints[1]),(editBottom[0],editBottom[1]),(0,0,255),1)
                
                editBottom.pop(1)
                editBottom.pop(0)
                
img =cv2.imread("img2.jpg")
cv2.namedWindow('ROI')
cv2.setMouseCallback('ROI',draw_rectangle)

while(1):
    cv2.imshow('ROI',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
    
```
